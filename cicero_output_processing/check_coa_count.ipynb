```python
infilename = '/nfs/lab/jnewsome/pbmc/cicero_v4/conns/pbmc1-15.b.cicero_conns.txt'

valDict = {}

infile = open(infilename)
count = 0
for line in infile:
    if not line.startswith('Peak'):
        s = line.split()
        if s[2] != 'NA':
            if float(s[2]) >= 0.5:
                
                key1 = s[0] + '__' + s[1]
                
                key1_back = s[1] + '__' + s[0]
                
                if not key1_back in valDict and not key1 in valDict:
                    valDict[key1] = s[2]
                count = count + 1
infile.close()
count
```
