```python
infilename = '/nfs/lab/jnewsome/references/gencode.v19.0kb_promoter.bed'
outfilename = '/nfs/lab/jnewsome/references/gencode.v19.0kb_promoter.20bpPad.10up10down.bed'
```

```python
infile = open(infilename)
outfile = open(outfilename, 'w')
for line in infile:
    line = line.strip()
    if len(line) > 0:
        s = line.split()
        chrom = s[0]
        start = int(s[1]) - 10
        end = int(s[2]) + 10
        name = s[3]
        outline = chrom + '\t' + str(start) + '\t' + str(end) + '\t' + name + '\n'
        outfile.write(outline)
infile.close()
outfile.close()
# chr1	69090	69091	OR4F5
```
