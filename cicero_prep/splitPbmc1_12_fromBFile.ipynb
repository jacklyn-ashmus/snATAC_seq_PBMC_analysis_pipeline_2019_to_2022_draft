```python
infilePrefix = '/nfs/lab/jnewsome/pbmc/cicero_v3/lft_mtx_origCode/pbmc1-15.b.'
infileSuffix = '.lft_mtx.mtx'
outfilePrefix = '/nfs/lab/jnewsome/pbmc/cicero_v3/lft_mtx_origCode/pbmc1_12/pbmc1_12.new.b.'

outfileSuffix = '.lft_mtx.mtx'
```

```python
samples = ['pbmc1', 'pbmc2','pbmc3','pbmc4','pbmc5','pbmc6','pbmc8','pbmc9','pbmc10','pbmc12']


```

```python
for i in range(1,23):
    chrom = "chr" + str(i) 
    infilename = infilePrefix + chrom + infileSuffix
    outfilename = outfilePrefix + chrom + outfileSuffix
    print(chrom)
    infile = open(infilename)
    outfile = open(outfilename, 'w')
    for line in infile:
        line = line.strip()
        if len(line) > 0:
            s = line.split()
            bar = s[1]
            # 11:10007826-10008046	pbmc10_CTTTGCGCATACAACC	60
            
            for samp in samples:
                sampthing = samp + '_'
                if bar.startswith(sampthing):
                    outfile.write(line + '\n')
    infile.close()
    outfile.close()
    
    
            
```
