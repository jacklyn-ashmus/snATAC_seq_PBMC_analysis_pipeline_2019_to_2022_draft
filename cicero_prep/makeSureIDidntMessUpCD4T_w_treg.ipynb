```python
def list2dict (list1):
    dict1 = {}
    for l in list1:
        dict1[l] = l
    return dict1

def readSingleColumnStringsFromFile (infilename):
    infile = open(infilename)
    list1 = list()
    for line in infile:
        line = line.strip()
        if len(line) > 0:
            list1.append(line)
    infile.close()
    return list1


def readBarcodesFromTagalign (infilename, fileSize):
    dict1 = {}
    infile = open(infilename)
    list1 = list()
    count = 1
    for line in infile:
        line = line.strip()
        if len(line) > 0:
            s = line.split()
            bar = s[3]
            dict1[bar] = ''
        if count % 1000000 == 0:
            if fileSize > 0:
                totalSize = fileSize / 1000000
                c = count / 1000000
                prog = c / totalSize
                print(prog, end = "\r")
            else:
                c = count / 1000000
                print(c, 'M', end = "\r")
        count = count + 1
    infile.close()
    return dict1
```
