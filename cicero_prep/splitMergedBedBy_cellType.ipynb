```python
infilename = '/nfs/lab/projects/pbmc_snATAC/analysis_v2/peaks/pbmc.sorted.merged.bed'

ct = {
    't' : ['act_cd4_t', 'naive_cd4_t', 'cyto_cd8_t', 'tReg', 'naive_cd8_t', 'mem_cd8_t', ],
    'cd4_t'  : ['act_cd4_t', 'naive_cd4_t',  'tReg', ],
    'cd8_t'  : [ 'cyto_cd8_t',  'naive_cd8_t', 'mem_cd8_t' ],
    'nk'  : [ 'adaptive_NK',  'cyto_nk' ],
    'mono'  : [ 'ncMono',  'cMono', 'iMono', 'cDC' ],
    'b'  : [ 'mem_b',  'naive_b' ],
    'plasma'  : ['plasma'],
    'mkc'  : ['mkc'],
    'pDC'  : ['pDC'],
    
    
    
    'act_cd4_t' : ['act_cd4_t' ],
    'naive_cd4_t' : ['naive_cd4_t', ],
    'cyto_cd8_t'  : ['cyto_cd8_t', ],
    'tReg'  : [ 'tReg',],
    'naive_cd8_t' : [ 'naive_cd8_t', ],
    'mem_cd8_t'  : ['mem_cd8_t', ],
    'adaptive_NK'  : [ 'adaptive_NK' ],
    'cyto_nk' : [ 'cyto_nk' ],
    
    'ncMono' : [ 'ncMono' ],
    'cMono'  : [ 'cMono' ],
    'iMono'  : [ 'iMono' ],
    'cDC' : ['cDC' ],
    
    'mem_b'  : [ 'mem_b' ],
    'naive_b' : ['naive_b' ],
    
    'cDC'  : ['cDC']
    
    
}





outfilePrefix = '/nfs/lab/jnewsome/pbmc/cicero_v4/split_merged_bed_byCellType/pbmc1_15.'
outfileSuffix =  '.mergedBare.cell_split.bed'

```

```python

for cell in ct.keys():
    print(cell)
    infile = open(infilename)
    outfilename = outfilePrefix + cell + outfileSuffix
    outfile = open(outfilename, 'w')
    
    for line in infile:
        line = line.strip()
        if len(line) > 0:
            s = line.split()
            chrom = s[0]
            start = s[1]
            end = s[2]
            
            ctList_str = s[3]
            
            ctList = ctList_str.split(',')
            
            found = False
            for ct1 in ctList:
                
                list1 = ct[cell]
                for ct2 in list1:
                    if ct2 in ct:
                        found = True
            
            if found:
                outline = chrom + '\t' + start + '\t' + end + '\n'
                outfile.write(outline)
                
    outfile.close()
    infile.close()

```
