```python
infilename = '/nfs/lab/jnewsome/pbmc/split_cicero_2/splitMeta/cellSplit/pbmc1-15.b.metadata.txt'
outfilename = '/nfs/lab/jnewsome/pbmc/split_cicero_2/splitMeta/cellSplit/pbmc1-12.b.metadata.txt'
samples = ['pbmc1', 'pbmc2','pbmc3','pbmc4','pbmc5','pbmc6','pbmc8','pbmc9','pbmc10','pbmc12']
```

```python

def subsetFile (metafilename, outfilename, samples):
    metaFile = open(metafilename)
    outfile = open(outfilename, 'w')
    for line in metaFile:
        line = line.strip()
        if len(line) > 0:
            if line.startswith('sample') or line.startswith('barcode'):
#                 outheaderline = 'barcode\t' + line + '\n'
                outfile.write(line + '\n')
            else:
                s = line.split('\t')
                barcode = s[0]
                for samp in samples:
                    sampThing = samp + '_'
                    if barcode.startswith(sampThing):
                        outfile.write(line + '\n')
    metaFile.close()
    outfile.close()    
    
```

```python
subsetFile (infilename, outfilename, samples)
```
