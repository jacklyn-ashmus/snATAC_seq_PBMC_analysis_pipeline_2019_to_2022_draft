```r
wd <- '/home/joshchiou/joshchiou-data2/islet_snATAC/fresh_only/cicero'
conns <- read.table(file.path(wd, 'alpha.cicero_coaccess_05.in_peak.conns.txt'), sep='\t', header=TRUE)
#wd <- '/home/joshchiou/joshchiou-data2/nPOD/merged_samples/cicero'
#conns <- read.table(file.path(wd, 'nPOD.ductal.cicero_conns.txt'), sep='\t', header=TRUE)
#wd <- '/home/joshchiou/joshchiou-data2/lung_snATAC/merged_samples/D110-3Y/cicero'
#conns <- read.table(file.path(wd, 'lung.b_cell.cicero_conns.txt'), sep='\t', header=TRUE)

#conns <- na.omit(conns)
#conns <- conns[conns$coaccess>=.05,]
rownames(conns) <- NULL

conns[c('chrA','startA','endA')] <- stringr::str_split_fixed(conns$Peak1, '_', 3)
conns[c('chrB','startB','endB')] <- stringr::str_split_fixed(conns$Peak2, '_', 3)
conns[,c('startA','endA','startB','endB')] <- sapply(conns[,c('startA','endA','startB','endB')], as.numeric)

comp <- read.table(file.path(wd, 'islet_pC-HiC.conns.txt'), sep='\t', header=TRUE)
comp[c('chrA','startA','endA')] <- stringr::str_split_fixed(comp$Peak1, '_', 3)
comp[c('chrB','startB','endB')] <- stringr::str_split_fixed(comp$Peak2, '_', 3)
comp[,c('startA','endA','startB','endB')] <- sapply(comp[,c('startA','endA','startB','endB')], as.numeric)

gene_annotation <- read.table('/home/joshchiou/references/gencode.v19.cicero_gene_annotation.txt', header=T, stringsAsFactors=FALSE)
gene_annotation <- gene_annotation[gene_annotation[['feature']]=='protein_coding',]
```

```r
peaks <- c('chr9_3936530_3937203','chr9_4299210_4300311')
#peaks <- c('chr7_44255624_44256447','chr7_44362212_44362831')

conns_subset <- conns[(conns$Peak1 %in% peaks) | (conns$Peak2 %in% peaks), ]
rownames(conns_subset) <- NULL
#chr7:14,900,000-15,200,000
chr = unlist(strsplit(peaks, '_'))[1]
pstart <- as.numeric(unlist(strsplit(peaks, '_'))[2])
pend <- as.numeric(unlist(strsplit(peaks, '_'))[3])
start = 3000000 #round(min(conns_subset$startA) - .01*(max(conns_subset$endA)-min(conns_subset$startA)))
end = 5000000 #round(max(conns_subset$endA) + .01*(max(conns_subset$endA)-min(conns_subset$startA)))

#conns_subset <- conns[(conns$chrA==chr) & (mapply(dplyr::between,conns$startA,start,end) | mapply(dplyr::between,conns$endA,start,end) | mapply(dplyr::between,conns$startB,start,end) | mapply(dplyr::between,conns$endB,start,end)),]
comp_subset <- comp[(comp$chrA==chr) & (mapply(dplyr::between,pstart,comp$startA,comp$endA) | mapply(dplyr::between,pend,comp$startA,comp$endA) | mapply(dplyr::between,pstart,comp$startB,comp$endB) | mapply(dplyr::between,pend,comp$startB,comp$endB)),]
```

```r
comp_subset
```

```r
ymax <- .25
conns_subset[conns_subset$coaccess>ymax, 'coaccess'] <- ymax

plot_list <- plot_connections(conns_subset, chr, start, end,
                 alpha_by_coaccess=TRUE,
                 coaccess_cutoff=.05,
                 connection_color='#008b00',
                 comparison_track=comp_subset,
                 comparison_connection_color='#92D5E6',
                 comparison_connection_width=1,
                 peak_color='black',
                 gene_model=gene_annotation,
                 gene_model_color='darkgrey',
                 connection_ymax=ymax,
                 connection_width=1,
                 collapseTranscripts='longest',
                 return_as_list=T)

pdf('/home/joshchiou/joshchiou-data2/islet_snATAC/pdf_plots/revision/cicero_analysis/fig_5.beta.GLIS3_locus.chr9_3000000-5000000.cicero.pdf')
Gviz::plotTracks(plot_list,  
                 sizes=c(.35,.1,.2,.35,.1,.05),
                 from=start, to=end, chromosome=chr, 
                 transcriptAnnotation=NULL,
                 col.axis='black',
                 fontsize=16,
                 fontsize.group=20,
                 fontcolor.legend='black',
                 lwd=.3,
                 title.width=.5,
                 background.title='transparent', 
                 col.border.title='transparent')
dev.off()
```
